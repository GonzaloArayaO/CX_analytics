```python
import pandas as pd
import tabula
```

```python
pdf_path = 'box_score/Planilla estad√≠stica FIBA CXO vs HAL 04 noviembre.pdf'
```

```python
tablas = tabula.read_pdf(pdf_path, pages="all", multiple_tables=True, encoding='ISO-8859-1')
```

```python
df = pd.DataFrame(tablas[1])
```

```python
df = df.drop(columns=["Unnamed: 1"])
```

```python
headers = ['NOMBRE', 'MIN', 'TC', 'T2', 'T3', 'T1', 'REB', 'AST', 'PER', 'REC', 'TAP', 'FALTAS', '+/-', 'PTS']
```

```python
df.columns = headers
```

```python
df = df.drop([0, 1])
```

```python
df = df[~df['NOMBRE'].str.contains("Equipo/Entrenador|Totales", na=False)]
```

```python
df['NOMBRE'] = df['NOMBRE'].str.replace(r'^\*?\d+\s*', '', regex=True)
```

```python
df['MIN'] = df['MIN'].replace('NJ', '0:0')
df[['MIN', 'SEG']] = df['MIN'].str.split(':', expand=True)
```

```python
df['TC'] = df['TC'].fillna('0/0')
df[['TCC', 'TCI']] = df['TC'].str.extract(r'(\d+)/(\d+)').fillna(0)
df = df.drop(columns=['TC'])
```

```python
df['T2'] = df['T2'].fillna('0/0')
df[['T2C', 'T2I']] = df['T2'].str.extract(r'(\d+)/(\d+)').fillna(0)
df = df.drop(columns=['T2'])
```

```python
df['T3'] = df['T3'].fillna('0/0')
df[['T3C', 'T3I']] = df['T3'].str.extract(r'(\d+)/(\d+)').fillna(0)
df = df.drop(columns=['T3'])
```

```python
df['T1'] = df['T1'].fillna('0/0')
df[['T1C', 'T1I']] = df['T1'].str.extract(r'(\d+)/(\d+)').fillna(0)
df = df.drop(columns=['T1'])
```

```python
df['REB'] = df['REB'].fillna('0 0 0')
df[['RO', 'RD', 'RT']] = df['REB'].str.split(' ', expand=True)
df = df.drop(columns=['REB'])
```

```python
df['FALTAS'] = df['FALTAS'].fillna('0 0')
df[['FALTA_CON', 'FALTA_REC']] = df['FALTAS'].str.split(' ', expand=True)
df = df.drop(columns=['FALTAS'])
```

```python
df['AST'] = df['AST'].fillna('0')
df['PER'] = df['PER'].fillna('0')
df['REC'] = df['REC'].fillna('0')
df['TAP'] = df['TAP'].fillna('0')
df['+/-'] = df['+/-'].fillna('0')
df['PTS'] = df['PTS'].fillna('0')
```
